```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = SparkSession.builder.master('local').appName('learning-df').config('spark.ui.port', '4050').getOrCreate()
```

```python
df1 = spark.read.format('csv').option('header', 'true').option('inferschema', 'true').option('delimiter', ';').load('/content/drive/MyDrive/datasets/arquivo_geral.csv')
```

```python
#WHEN - OTHERWISE
'''Criar uma nova coluna com as eguintes condições:
  Caso o numero de casos novos seja maior do que zero, colocar o texto tem casos novos
  Caso o numero de casos novos nao seja maior do que zero, colocar o texto sem casos novos
  '''
```

```python
df2 = (df1.withColumn('status', F.when(F.col('casosNovos')>0, 'Tem casos novos')
.otherwise('Nao tem casos novos')))
```

```python
df2.printSchema()
```

```python
df2.show(5)
```

```python
df2.select(F.col('regiao'), F.col('estado'), F.col('casosNovos'), F.col('status')).show(50)
```

```python
#passando a quantidade de casos novos e concatenado com o texto casos novos
df2.show(5)
```

```python
df3 = (df1.withColumn("status", 
          F.when(F.col("casosNovos") > 0, F.concat(df1.casosNovos,F.lit(" casos novos")))
          .otherwise("Não tem casos novos")))
```

```python
df3.select(F.col('data'), F.col('casosNovos'), F.col('status')).show(250)
```

```python
'''
Criando colunas com Dia/Mês/ANO para ficar mais organizado no DataFrame
Isso é feito com a utilizado da withColumn que cria uma nova coluna junto com a substring que mostra uma
parte da string de acordo com os parâmetros escolhidos

'''
```

```python
df1.select('data').show(5)
```

```python
df4 = df1.withColumn('ano', F.substring('data', 1, 4)).withColumn('mes', F.substring('data', 6, 2)).withColumn('dia', F.substring('data', 9, 2))
```

```python
df4.show(5)
```

```python
df4.select(F.col('casosNovos'), F.col('dia'), F.col('mes'), F.col('ano')).show(50)
```

```python
df5 =(
    df1.withColumn('Status',
    F.when(F.col('casosNovos')> 0, F.concat(df1.casosNovos, F.lit(' caso(s) novo(s)')))
    .otherwise('Sem casos novos')).withColumn('ano', F.substring('data', 1, 4))
    .withColumn('mes', F.substring('data', 6, 2))
    .withColumn('dia', F.substring('data', 9, 2))

)
```
