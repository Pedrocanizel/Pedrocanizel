```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
spark = SparkSession.builder.master('local').appName('learning-df').config('spark.ui.port', '4050').getOrCreate()
```

```python
df1 = spark.read.format('csv').option('header', 'false').option('inferschema', 'true').option('delimiter', ';').load('/content/drive/MyDrive/datasets/arquivo_geral.csv').toDF("UM", "DOIS", "TRES", "QUATRO", "CINCO", "SEIS", "SETE")
```

```python
df1.printSchema()
```

```python
#segunda maneira
lista_esquema = ['UM', 'DOIS', 'TRES', 'QUATRO', 'CINCO', 'SEIS', 'SETE']
df1 = (spark
       .read
       .format("csv")
       .option("header", "false")
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/datasets/arquivo_geral.csv').toDF(*lista_esquema)
)
```

```python
df1.printSchema()
```

```python
#TERCEIRA MANEIRA - UTILIZANDO O SELECT COM F.COL E ALIAS
df2 = (spark
       .read
       .format("csv")
       .option("header", "false")
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/datasets/arquivo_geral.csv')
)
```

```python
df2 = df2.select(F.col("_c0").alias("um"),
                 F.col("_c1").alias("dois"),
                 F.col("_c2").alias("tres"),
                 F.col("_c3").alias("quatro"),
                 F.col("_c4").alias("cinco"),
                 F.col("_c5").alias("seis"),
                 F.col("_c6").alias("sete"),
)

df2.printSchema()
```
